## Player Financials


```python
# load and read csv file into DataFrame

financials_to_load = "Resources/NFL_Yearly_Breakdown_combined.csv"
financials_df = pd.read_csv(financials_to_load)
financials_df
```

```python
# Look for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Drop columns

financials_df = financials_df.drop(columns=["Rank","Unnamed: 19","Age", "FA Year", 
        "Sign Year", "Sign Age","Cash","Draft Year","AAV","CONT YR","CONT VALUE","Earnings"])
financials_df
```

```python
# Drop rows with future years

values = [2022, 2023, 2024]
financials_df = financials_df[financials_df.Year.isin(values)==False]
financials_df
```

```python
# Check again for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Combine the records of renamed teams; look at random selection

financials_df["Team"].replace({"LAR":"LAR/STL", "STL":"LAR/STL", "JAC":"JAX","LV": "LV/OAK","OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
financials_df.sample(25)
```

```python
# Export financial Dataframe to csv

financials_df_path = "Resources/financials_df.csv"
financials_df.to_csv(financials_df_path)
```

## Team Standings

```python
# load and read csv file into DataFrame

standings_to_load = "Resources/nfl_standings.csv"
nfl_standings_df = pd.read_csv(standings_to_load)
nfl_standings_df
```

```python
# Change column headings (to be consistent with financial dataframe)

nfl_standings_df = nfl_standings_df.rename(
    columns={"season": "Year", "pct": "Win %","conf":"Conf","division":"Division",
    "team":"Team","wins":"Wins","losses":"Losses","ties":"Ties","div_rank":"Div_Rank",
    "scored":"Scored","allowed":"Allowed","net":"Net","sov":"SOV","sos":"SOS","playoff":"Playoff"})
nfl_standings_df
```

```python
# Combine records of renamed teams

nfl_standings_df["Team"].replace({"LA": "LAR/STL", "STL": "LAR/STL", "LV": "LV/OAK",
                                 "OAK": "LV/OAK", "SD": "LAC/SD", "LAC": "LAC/SD"}, inplace=True),
nfl_standings_df
```

```python
# Drop data (rows) prior to 2011,

nfl_standings_df.drop(
    nfl_standings_df.loc[nfl_standings_df["Year"] < 2011].index, inplace=True)
nfl_standings_df
```

```python
# Add column for superbowl champions

# create conditions for superbowl win for teams that made playoffs
superbowl_conditions = [
    (nfl_standings_df["Playoff"] == "WonSB"),
    (nfl_standings_df["Playoff"] == "LostSB"),
    (nfl_standings_df["Playoff"] == "LostCC"),
    (nfl_standings_df["Playoff"] == "LostDV"),
    (nfl_standings_df["Playoff"] == "LostWC")
]

# Create a list of the values to assign for each condition
values = ["Yes", "No", "No", "No", "No"]

# create a new column and use np.select to assign values for winning column
nfl_standings_df["Won Superbowl"] = np.select(superbowl_conditions, values)

# replace null values with No for teams that did not make playoffs
nfl_standings_df["Won Superbowl"].replace({"0": "No"}, inplace=True)

# display 1 year of data to ensure 1 champion
nfl_standings_df.head(32)
```

```python
# change playoff column to Yes or No

nfl_standings_df["Playoff"].replace({"LostSB": "Yes", "LostCC": "Yes", "LostDV": "Yes",
                                     "LostWC": "Yes", "WonSB": "Yes"}, inplace=True)

nfl_standings_df["Playoff"].fillna("No", inplace=True)

nfl_standings_df
```

```python
# Drop column

nfl_standings_df = nfl_standings_df.drop(columns="seed")
nfl_standings_df
```

```python
# Export team standings Dataframe to csv

nfl_standings_df_path = "Resources/standings_df.csv"
nfl_standings_df.to_csv(nfl_standings_df_path)
```
