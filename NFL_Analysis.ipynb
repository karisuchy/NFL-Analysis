```python
# Import Dependencies

import pandas as pd
import numpy as np
```

## Player Financials


```python
# load and read csv file into DataFrame

financials_to_load = "Resources/NFL_Yearly_Breakdown_combined.csv"
financials_df = pd.read_csv(financials_to_load)
financials_df
```

```python
# Look for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Drop columns

financials_df = financials_df.drop(columns=["Rank","Unnamed: 19","Age", "FA Year", 
        "Sign Year", "Sign Age","Cash","Draft Year","AAV","CONT YR","CONT VALUE","Earnings"])
financials_df
```

```python
# Drop rows with future years

values = [2022, 2023, 2024]
financials_df = financials_df[financials_df.Year.isin(values)==False]
financials_df
```

```python
# Check again for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Combine the records of renamed teams; look at random selection

financials_df["Team"].replace({"LAR":"LAR/STL", "STL":"LAR/STL", "JAC":"JAX","LV": "LV/OAK","OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
financials_df.sample(25)
```

```python
# export financial df to csv

financials_df_path = "Resources/financials_df.csv"
financials_df.to_csv(financials_df_path)
```
